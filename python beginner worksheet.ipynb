```python
#write a programe that solves a quadratic equation


#accepting users input
a = int(input('Enter your value for a: '))
b = int(input('Enter your value for b: '))
c = int(input('Enter your value for c: '))

#define the solution

x1 = (-b + ((b**2) - (4*a*c)) ** (1/2)) /(2*a)
x2 = (-b - ((b**2) - (4*a*c)) ** (1/2)) /(2*a) 


print(f'The root of the equation is {x1} and {x2}')
```

```python
def read_query(connection, query):
    result = None
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        result = cursor.fetchall()
    except Error as err:
        print(f"Error: {err}")
    
    return result
```

```python
def count_vowels(name):
    count = []
    vowel = ['a','e','i','o','u']
    for letter in name:
        if letter in vowel and letter not in count:
            count.append(letter)
        else:
            pass
    return len(count)

count_vowels('awele')
```
